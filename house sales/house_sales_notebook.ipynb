```python
import math
df['city'].replace('--', 'Unknown', inplace=True)
df['house_type'].replace('Terr.', 'Terraced', inplace=True)
df['house_type'].replace('Semi', 'Semi-detached', inplace=True)
df['house_type'].replace('Det.', 'Detached', inplace= True)
df["months_listed"].fillna(df["months_listed"].mean(), inplace=True)
df["months_listed"] = df["months_listed"].apply(lambda x: math.ceil(x * 10) / 10)
clean_data = df
clean_data
```

```python
price_by_rooms = df.groupby("bedrooms").agg(avg_price=("sale_price", "mean"), var_price=("sale_price", "var")).round(1).reset_index()
price_by_rooms
```

```python
train_df = pd.read_csv("train.csv")
train_df['city'].replace('Silvertown', '1', inplace=True)
train_df['city'].replace('Riverford', '2', inplace=True)
train_df['city'].replace('Teasdale', '3', inplace=True)
train_df['city'].replace('Poppleton', '4', inplace=True)
train_df['house_type'].replace('Terraced', '1', inplace=True)
train_df['house_type'].replace('Semi-detached', '2', inplace=True)
train_df['house_type'].replace('Detached', '3', inplace= True)
train_df
```

```python

validation_df = pd.read_csv("validation.csv")
validation_df['city'].replace('Silvertown', '1', inplace=True)
validation_df['city'].replace('Riverford', '2', inplace=True)
validation_df['city'].replace('Teasdale', '3', inplace=True)
validation_df['city'].replace('Poppleton', '4', inplace=True)
validation_df['house_type'].replace('Terraced', '1', inplace=True)
validation_df['house_type'].replace('Semi-detached', '2', inplace=True)
validation_df['house_type'].replace('Detached', '3', inplace= True)
validation_df
```

```python
# Import necessary libraries
# import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the training and validation datasets
train_data = train_df
validation_data = validation_df

# Feature selection
features = ['bedrooms', 'house_type', 'area', 'city']

# Extract features and target variable from the training data
X_train = train_data[features]
y_train = train_data['sale_price']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the validation data
X_validation = validation_data[features]
base_predictions = model.predict(X_validation)

# Create a dataframe with house_id and predicted price
base_result = pd.DataFrame({'house_id': validation_data['house_id'], 'price': base_predictions})

# Display the result
print(base_result.head())

# Save the result to a CSV file
base_result.to_csv('base_result.csv', index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Load the training and validation datasets
train_data = train_df
validation_data = validation_df

# Feature selection
features = ['bedrooms', 'house_type', 'area', 'city']

# Extract features and target variable from the training data
X_train = train_data[features]
y_train = train_data['sale_price']

# Create and fit the Random Forest Regressor model
model_compare = RandomForestRegressor(random_state=42)
model_compare.fit(X_train, y_train)

# Make predictions on the validation data
compare_predictions = model_compare.predict(X_validation)

# Create a dataframe with house_id and predicted price
compare_result = pd.DataFrame({'house_id': validation_data['house_id'], 'price': compare_predictions})

# Display the result
print(compare_result.head())

# Save the result to a CSV file
compare_result.to_csv('compare_result.csv', index=False)

```
